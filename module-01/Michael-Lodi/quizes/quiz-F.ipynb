```python
extra_day = 40
defined_days = {
    1: 39.99,
    2: 79.99,
    3: 120,
    4: 160,
}

def rent(days: int) -> float:
    if days in defined_days:
        return float(defined_days[days]) 
    else:
        return float(defined_days[3] + (days - 3) * extra_day) if days > 0 else float(0)
    
data = [-2, 0, 1, 2, 3, 4, 5, 10]

for d in data:
    print(rent(d))
```

            <td>True</td>
        </tr>
        <tr>

            <td>True</td>
        </tr>
        <tr>
            <td>print(length_multiple("Python", 0))</td>

        </tr>
        <tr>
            <td>print(length_multiple("", 0))</td>

        </tr>
        <tr>
            <td>print(length_multiple("", 42))</td>
            <td>True</td>

            <td>True</td>
        </tr>
    </tbody>

```python
import pandas as pd

def length_multiple(s: str, n: int) -> bool:
    return len(s) % n == 0 if n != 0 else False

data = [
    ("Cremona", 5, False),
    ("Cremona", 7, True),
    ("Python", 3, True),
    ("Cremona", 1, True),
    ("Python", 0, False),
    ("", 0, False),
    ("", 42, True),
    ("Ciao", -2, True),
]

de = pd.DataFrame(data, columns=["s", "n", "expected"])
de["result"] = de.apply(lambda x: length_multiple(x["s"], x["n"]), axis=1)
de["accurate"] = de["expected"] == de["result"]
print(de)
```

<h2 style="font-family: 'Courier New', Courier, monospace; font-size:40px;">

<p style="font-family: 'Courier New', Courier, monospace;">
Write a function solve_linear(a,b) that takes two floats a and b and solves the linear equation ax+b=0.
</p>
<p style="font-family: 'Courier New', Courier, monospace;">
If the equation is inconsistent, return the string inconsistent, while if it has infinite solutions, the string infinite
</p>
<p style="font-family: 'Courier New', Courier, monospace;">
For example:
</p>
<table style="font-family: 'Courier New', Courier, monospace;">
    <thead>
        <tr>
            <th>Test</th>
            <th>Result</th>
        </tr>
    </thead>
    <tbody>
        <tr>
            <td>print(solve_linear(5,2.5))</td>
            <td>-0.5</td>
        </tr>
        <tr>
            <td>print(solve_linear(0,0))</td>
            <td>infinite</td>
        </tr>
        <tr>
            <td>print(solve_linear(0,3.14))</td>
            <td>inconsistent</td>
        </tr>
    </tbody>
</table>


```python
from typing import Union

def solve_linear_possible(a: float, b: float) -> float:
    return -b / a

def solve_linear_infinite(a: float, b: float) -> str:
    return "infinite"

def solve_linear_inconsistent(a: float, b: float) -> str:
    return "inconsistent"

def solve_linear(a: float, b: float) -> Union[float, str]:
    # ax + b = 0
    # x = -b/a
    # Possible returns:
    # 1. If possible (5, 2.5): return the solution (-0.5)
    # 2. If infinite solutions (0, 0): return int("infinite")
    # 3. If is inconsistent (0, 3.14): return int("inconsistent")
    if a == 0 and b == 0:
        return solve_linear_infinite(a, b)
    elif a == 0:
        return solve_linear_inconsistent(a, b)
    else:
        return solve_linear_possible(a, b)
    
data = [
    (5, 2.5, -0.5),
    (0, 0, "infinite"),
    (0, 3.14, "inconsistent"),
    (0, 0, "infinite"),
    (1, 0, 0),
    (1, 1, -1),
    (2, 3, -1.5),
    (3, 2, -2/3),
]

de = pd.DataFrame(data, columns=["a", "b", "expected"])
de["result"] = de.apply(lambda x: solve_linear(x["a"], x["b"]), axis=1)
de["accurate"] = de["expected"] == de["result"]
print(de)
```

<p style="font-family: 'Courier New', Courier, monospace;">
Write a function solve_quadratic(a,b,c) that takes three floats a, b and c and solves (for real solutions only) the quadratic equation ax<sup>2</sup>+bx+c=0. The function returns a float if there is one solution, two floats for two solutions, and a string in the other cases:
</p>
<p style="font-family: 'Courier New', Courier, monospace;">
If the equation is inconsistent, return the string inconsistent, while if it has infinite solutions, the string infinite. If the discriminant b<sup>2</sup>âˆ’4ac&lt;0, returns the string complex.
</p>
<p style="font-family: 'Courier New', Courier, monospace;">
If a=0, the program should still work to solve the linear equation bx+c=0.
</p>
<p style="font-family: 'Courier New', Courier, monospace;">
For example:
</p>
<table style="font-family: 'Courier New', Courier, monospace;">
    <thead>
        <tr>
            <th>Test</th>
            <th>Result</th>
        </tr>
    </thead>
    <tbody>
        <tr>
            <td>print(solve_quadratic(0,0,0))</td>
            <td>infinite</td>
        </tr>
        <tr>
            <td>print(solve_quadratic(0,0,3))</td>
            <td>inconsistent</td>
        </tr>
        <tr>
            <td>print(solve_quadratic(0,-3.5,7))</td>
            <td>2.0</td>
        </tr>
        <tr>
            <td>print(solve_quadratic(1,2,3))</td>
            <td>complex</td>
        </tr>
        <tr>
            <td>print(solve_quadratic(5,6,1))</td>
            <td>(-1.0, -0.2)</td>
        </tr>
    </tbody>
</table>


```python
# a * x ** 2 + b * x + c = 0
from typing import Union

def solve_linear_possible(a: float, b: float) -> float:
    return -b / a

def solve_linear_infinite(a: float, b: float) -> str:
    return "infinite"

def solve_linear_inconsistent(a: float, b: float) -> str:
    return "inconsistent"

def solve_linear(a: float, b: float) -> Union[float, str]:
    if a == 0 and b == 0:
        return solve_linear_infinite(a, b)
    elif a == 0:
        return solve_linear_inconsistent(a, b)
    else:
        return solve_linear_possible(a, b)

def solve_quadratic_delta(a: float, b: float, c: float) -> float:
    return b ** 2 - 4 * a * c

def solve_quadratic_complex(a: float, b: float, c: float) -> float:
    return "complex"

def solve_quadratic(a: float, b: float, c: float) -> Union[float, str]:
    if a == 0:
        return solve_linear(b, c)
    else:
        delta = solve_quadratic_delta(a, b, c)
        if delta < 0:
            return solve_quadratic_complex(a, b, c)
        elif delta == 0:
            return float(-b / (2 * a))
        else:
            x1 = (-b - delta ** 0.5) / (2 * a)
            x2 = (-b + delta ** 0.5) / (2 * a)
            return float(x1), float(x2)


data = [
    (0, 0, 0, "infinite"),
    (0, 0, 3, "inconsistent"),
    (0, -3.5, 7, 2.0),
    (1, 2, 3, "complex"),
    (5, 6, 1, (-1.0, -0.2)),
]

de = pd.DataFrame(data, columns=["a", "b", "c", "expected"])
de["result"] = de.apply(lambda x: solve_quadratic(x["a"], x["b"], x["c"]), axis=1)
de["accurate"] = de["expected"] == de["result"]
print(de)
```

    Write a function ticket(age) that takes an integer parameter representing an age and returns a string describing the ticket corresponding to that age.

<p style="font-family: 'Courier New', Courier, monospace;">
    "Free" if age is less than 6<br>

    "Full ticket" if age is between 12 and 60 (endpoints excluded)<br>
    "Retirement discount" if age is 60 or more, but no more than 70<br>
    "Free" if age is strictly greater than 70

        </tr>
        <tr>
            <td>print(ticket(12))</td>
            <td>Child discount</td>
        </tr>
        <tr>
            <td>print(ticket(13))</td>

        </tr>

            <td>Full ticket</td>

            <td>print(ticket(70))</td>
            <td>Retirement discount</td>

            <td>print(ticket(71))</td>
            <td>Free</td>
        </tr>
    </tbody>

```python
import pandas as pd

def ticket(age: int) -> str:
    if age < 6:
        return "Free"
    elif age <= 12:
        return "Child discount"
    elif age < 60:
        return "Full ticket"
    elif age <= 70:
        return "Retirement discount"
    else:
        return "Free"
    
data = [
    (0, "Free"),
    (5, "Free"),
    (6, "Child discount"),
    (12, "Child discount"),
    (13, "Full ticket"),
    (59, "Full ticket"),
    (60, "Retirement discount"),
    (69, "Retirement discount"),
    (70, "Retirement discount"),
    (71, "Free"),
    (100, "Free"),
]

de = pd.DataFrame(data, columns=["age", "expected"])
de["result"] = de.apply(lambda x: ticket(x["age"]), axis=1)
de["accurate"] = de["expected"] == de["result"]
print(de)
```

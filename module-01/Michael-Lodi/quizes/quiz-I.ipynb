<p style="font-family: 'Courier New', Courier, monospace;">
Write a function palindrome(s) that returns True if and only if the string s is a palindrome (the same string if read from left to right and from right to left).
Suppose - without checking - that s is composed only by digits and lowercase letters (no punctuation, whitespace, special characters).
</p>


<p style="font-family: 'Courier New', Courier, monospace;">
Try to solve the problem both by using double slices [::], and by not using them.
</p>

<p style="font-family: 'Courier New', Courier, monospace;">
For example:
</p>

<table style="font-family: 'Courier New', Courier, monospace;">

            <th>Test</th>
            <th>Result</th>
        </tr>
    </thead>
    <tbody>
        <tr>
            <td>print(palindrome('ada'))</td>
            <td>True</td>
        </tr>
        <tr>
            <td>print(palindrome('nada'))</td>
            <td>False</td>
        </tr>
        <tr>
            <td>print(palindrome(('otto')))</td>
            <td>True</td>
        </tr>
        <tr>
            <td>print(palindrome('8'))</td>
            <td>True</td>
        </tr>
        <tr>
            <td>print(palindrome(''))</td>
            <td>True</td>
        </tr>
        <tr>
            <td>print(palindrome('232'))</td>
            <td>True</td>
        </tr>
        <tr>
            <td>print(palindrome('0232'))</td>
            <td>False</td>
        </tr>
    </tbody>
</table>



```python
# Question 01:

def palindrome(s: str) -> bool:
    return s[:] == s[::-1]


data = [
    {
        "value" : "ada",
        "answer": True,
    },
    {
        "value" : "nada",
        "answer": False,
    },
    {
        "value" : "otto",
        "answer": True,
    },
    {
        "value" : "8",
        "answer": True,
    },
    {
        "value" : "",
        "answer": True,
    },
    {
        "value" : "232",
        "answer": True,
    },
    {
        "value" : "0232",
        "answer": False,
    },
]

for d in data:
    print(palindrome(d["value"]) == d["answer"])

```

<p style="font-family: 'Courier New', Courier, monospace;">
Write a function even_couples(T) that, given a tuple T of integers, returns True if and only if all pairs of consecutive elements of T have an even sum. That means that only one counterexample (two adjacent elements of T whose sum is odd) is enough for the function to return False.
</p>

<p style="font-family: 'Courier New', Courier, monospace;">
If T is not a tuple, the function returns the string 'error'.<br>
If T has less than two elements, the function returns the string 'short'.<br>
Suppose - without checking - that all the elements of T are int numbers.
</p>

<p style="font-family: 'Courier New', Courier, monospace;">
Try to solve the exercise both by using while, and by using for.

```python
# Question 02:

from typing import Union

def even_couples(T: tuple) -> Union[bool, str]:
    if not isinstance(T, tuple):
        return "error"
    
    T_size = len(T)
    
    if T_size < 2:
        return "short"

    for indice in range(T_size - 1):
        sum_numbers = T[indice] + T[indice + 1]
        if sum_numbers % 2 != 0:
            return False
        
    return True

data = [
    {
        "value" : (6, 2, 0, 4),
        "answer": True,
    },
    {
        "value" : (6, 2, 0, 4, 1),
        "answer": False,
    },
    {
        "value" : (1, 3, 11, 7),
        "answer": True,
    },
    {
        "value" : (1, 3, 12, 7),
        "answer": False,
    },
    {
        "value" : (10, ),
        "answer": "short",
    },
    {
        "value" : 10,
        "answer": "error",
    },
]

for d in data:
    print( even_couples(d["value"]) == d["answer"]  )
```

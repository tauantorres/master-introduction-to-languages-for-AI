```python
# Question 01:
def greetings(msg: str, name: str) -> str:
    return f'{msg}, {name}!' if isinstance(name, str) and isinstance(msg, str) else 'bad params'


data = [
    ('hello', 'Marco', 'hello, Marco!'),
    ("Salut", "Lady Jessica", "Salut, Lady Jessica!"),
    ("Salut", 42, "bad params"),
]

for msg, name, expected in data:
    result = greetings(msg, name)
    print(result == expected, result)
```

<p style="font-family: 'Courier New', Courier, monospace;">
Write a function circumference(sq_side) that takes a numeric parameter representing the side of a square inscribed in a circle and returns the circumference of that circle.
The function returns None if the square side is negative.
</p>

<p style="font-family: 'Courier New', Courier, monospace;">
To write the function you should use:
the given name pi - imported from the math library in the first line - for the value of pi
the given function square_diagonal(side), to get the diagonal of a square from its side
Be sure to define the function name exactly as requested, or the automatic tests will fail.
</p>

<p style="font-family: 'Courier New', Courier, monospace;">
A red square inscribed in a blue circle
P.S. The round function in some of the tests approximates the result to 2 decimal digits, to prevent correct solutions from failing because of the inaccuracy of floating point calculations.
</p>

<p style="font-family: 'Courier New', Courier, monospace;">
For example:
</p>

<table style="font-family: 'Courier New', Courier, monospace;">
    <thead>
        <tr>
            <th>Test</th>
            <th>Result</th>
        </tr>
    </thead>
    <tbody>
        <tr>
            <td>print(circumference(-1))</td>
            <td>None</td>
        </tr>
        <tr>
            <td>print(circumference(0))</td>
            <td>0.0</td>
        </tr>
        <tr>
            <td>print(round(circumference(3), 2))</td>
            <td>13.33</td>
        </tr>
        <tr>
            <td>print(round(circumference(10), 2))</td>
            <td>44.43</td>
        </tr>
    </tbody>
</table>


```python
from typing import Union
from math import pi, sqrt

def square_diagonal(side: float) -> Union[float, None]:
    if side < 0:
        return None
    return side * sqrt(2)

def circumference(sq_side: float) -> Union[float, None]:
    diagonal = square_diagonal(sq_side)
    if diagonal is None:
        return None
    return 2 * pi * (diagonal / 2)


data = [
    (-1, None),
    (0, 0.0),
    (3, 13.33),
    (10, 44.43)
]

for d in data:
    side, expected = d
    result = circumference(side)
    if result is not None:
        result = round(result, 2)
    print(result == expected, result)
```

Write a function ticket(age) that takes an integer parameter representing an age, and returns a string describing the ticket corresponding to that age, as follows:
</p>

<p style="font-family: 'Courier New', Courier, monospace;">
"Free" if age is less than 6
"Child discount" if age is less than or equal to 12
"Full ticket" if age is between 12 and 60 (endpoints excluded)
"Retirement discount" if age is 60 or more
"Free" if age is strictly greater than 70
Be sure to define the function name exactly as requested, or the automatic tests will fail.
</p>

```python
# Question 03:

def ticket(age: int) -> str:
    if age < 6:
        return "Free"
    elif age <= 12:
        return "Child discount"
    elif age < 60:
        return "Full ticket"
    elif age <= 70:
        return "Retirement discount"
    else:
        return "Free"
    
data = [
    (0, "Free"),
    (5, "Free"),
    (6, "Child discount"),
    (12, "Child discount"),
    (13, "Full ticket"),
    (59, "Full ticket"),
    (60, "Retirement discount"),
    (70, "Retirement discount"),
    (71, "Free")
]

for age, expected in data:
    result = ticket(age)
    print(result == expected, result)
```

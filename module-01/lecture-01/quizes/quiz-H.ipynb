<p style="font-family: 'Courier New', Courier, monospace;">
Write a function digits_tuple(num) that takes an integer num and returns the tuple whose elements are the digits (and the sign, if present) of num.
Suppose - without checking - that num is an integer number.
</p>

<p style="font-family: 'Courier New', Courier, monospace;">
For example:
</p>
<table style="font-family: 'Courier New', Courier, monospace;">
    <thead>
        <tr>
            <th>Test</th>
            <th>Result</th>
        </tr>
    </thead>
    <tbody>
        <tr>
            <td>print(digits_tuple(125))</td>
            <td>('1', '2', '5')</td>
        </tr>
        <tr>
            <td>print(digits_tuple(-125))</td>
            <td>('-', '1', '2', '5')</td>
        </tr>
        <tr>
            <td>print(digits_tuple(0))</td>
            <td>('0',)</td>
        </tr>
        <tr>
            <td>print(digits_tuple(7))</td>
            <td>('7',)</td>
        </tr>
    </tbody>
</table>

```python
# Question 01:

def digits_tuple(num: int) -> tuple:
    numbers_in_str = str(num)
    return tuple(number_in_str for number_in_str in numbers_in_str)

data = [125, -125, 0, 7]
for num in data:
    print(digits_tuple(num))
```

<p style="font-family: 'Courier New', Courier, monospace;">
Write the function mul_integers(N) that takes the tuple N, multiply all the int elements in N, and returns the product.
If N is not a tuple, the functions returns the int number -999.
If N is a tuple, but does not contain any int number, the function returns 1.
Scanning the tuple N, when a non-integer element, or the int number 0 is encountered, that element is ignored.
</p>

<p style="font-family: 'Courier New', Courier, monospace;">
For example:
</p>
<table style="font-family: 'Courier New', Courier, monospace;">
    <thead>
        <tr>
            <th>Test</th>
            <th>Result</th>
        </tr>
    </thead>
    <tbody>
        <tr>
            <td>print(mul_integers((1,2,3,1)))</td>
            <td>6</td>
        </tr>
        <tr>
            <td>print(mul_integers((1,2,3,1,0)))</td>
            <td>6</td>
        </tr>
        <tr>
            <td>print(mul_integers((1,2,3,1,(1,2))))</td>
            <td>6</td>
        </tr>
        <tr>
            <td>print(mul_integers((1,2,3,1,2.0)))</td>
            <td>6</td>
        </tr>
        <tr>
            <td>print(mul_integers(('hi',1,2,3,1)))</td>
            <td>6</td>
        </tr>
        <tr>
            <td>print(mul_integers((2.0,'house',3.3)))</td>
            <td>1</td>
        </tr>
        <tr>
            <td>print(mul_integers(()))</td>
            <td>1</td>
        </tr>
        <tr>
            <td>print(mul_integers("cat"))</td>
            <td>-999</td>
        </tr>
        <tr>
            <td>print(mul_integers(1231))</td>
            <td>-999</td>
        </tr>
    </tbody>
</table>

```python
# Questione 02:

def mul_integers(N: tuple, initial_product: int = 1) -> int:

    if not isinstance(N, tuple):
        return -999

    if not any(isinstance(num, int) for num in N):
        return 1

    for num in N:
        if isinstance(num, int) and num != 0:
            initial_product *= num
    return initial_product


data = [
    (1, 2, 3, 1),
    (1, 2, 3, 1, 0),
    (1, 2, 3, 1, (1, 2)),
    (1, 2, 3, 1, 2.0),
    ("hi", 1, 2, 3, 1),
    (2.0, "house", 3.3),
    (),
    ("cat"),
    1231
]

for d in data:
    print(mul_integers(d))
```

<p style="font-family: 'Courier New', Courier, monospace;">
Write a function get_vowels(S) that returns a tuple with all the unique vowels found in the string S.
The result tuple contains the vowels in the same order they are present in S.
If a vowel is found in S more than once, only the first occurrence is considered.
Suppose - without checking - that S is composed only by lowercase characters.
</p>

<p style="font-family: 'Courier New', Courier, monospace;">
Tips:
<ul>
    <li>returning a tuple is different than printing</li>
    <li>the result tuple should be initialized to the empty tuple before the linear scan of S</li>
    <li>see slide 20 on how to "add" elements to a tuple</li>
    <li>see slide 18 ("Examples vs. non examples") on how to express a tuple of a single element</li>
</ul>
</p>

<p style="font-family: 'Courier New', Courier, monospace;">
For example:
</p>
<table style="font-family: 'Courier New', Courier, monospace;">
    <thead>
        <tr>
            <th>Test</th>
            <th>Result</th>
        </tr>
    </thead>
    <tbody>
        <tr>
            <td>print(get_vowels("ciao"))</td>
            <td>('i', 'a', 'o')</td>
        </tr>
        <tr>
            <td>print(get_vowels("amarcord"))</td>
            <td>('a', 'o')</td>
        </tr>
        <tr>
            <td>print(get_vowels("xyz_123"))</td>
            <td>()</td>
        </tr>
        <tr>
            <td>print(get_vowels(""))</td>
            <td>()</td>
        </tr>
    </tbody>
</table>

```python
# Question 03:
def get_vowels(S: str)-> tuple:
    vowels = ["a", "e", "i", "o", "u"]
    S_list = list(S)
    S_return = []
    for s in S_list:
        if (s in vowels) and (s not in S_return):
            S_return.append(s)

    return tuple(S_return)

data = [
    {
        "value" : ("ciao"), 
        "answer": ("i", "a", "o")
    },
    {
        "value" : ("amarcord"), 
        "answer": ("a", "o")
    },
    {
        "value" : ("xyz_123"), 
        "answer": ()
    },
    {
        "value" : (""), 
        "answer": ()
    },
]

for d in data:
    print( get_vowels(d["value"]) == d["answer"])
```

Question 01:
Write a function add_n_print(L,obj,pos,replace) that takes  4 parameters:
L: a list
obj: any object to be added to the list
pos: the position in the list where to add the object
replace: a boolean value (True|False) indicating if the object replaces the previous one at index pos
Ideally, the function returns L with obj added at index pos.
In some cases, the function returns an error message.

Consider that:
if L is not a list, the function returns the string "not a list".
if pos is not a valid index of the list, the function returns "invalid pos"; note that, if pos==len(L) (i.e., the first position after the last valid index) the object is added after the last one.
if replace is False, the (possible) objects after pos shift right by one.
For example:

Test	Result
print(add_n_print(99,7,0,False))
not a list
print(add_n_print([99],7,0,False))
[7, 99]
print(add_n_print([99],7,1,False))
[99, 7]
print(add_n_print([99],7,2,False))
invalid pos
print(add_n_print([1,2,3],'a',2,False))
[1, 2, 'a', 3]
print(add_n_print([1,2,3],'a',2,True))
[1, 2, 'a']
print(add_n_print([1,2,3],'a',3,False))
[1, 2, 3, 'a']

Question 02:

Write a function str_to_list(s) that takes as parameter a string s and returns a list with some of the characters of the input string.

If all the characters in s are alphabetic, the result list contains only the consonants, in the same order they appear in the string. Whether the consonants in s are lower or upper case letters, they are put in the list capitalized.
If all the characters in s are digits, the result list contains only the odd digits, but in reverse order.
NB: The result list contains unique characters: if a character (consonant or odd digit) appears in s more than once, only the first occurrence is considered.

Special cases:
If s is not a string, the function returns 'not a string'.
If s does not contains only alphabetic characters, or only digits, the function returns 'alphanumeric'.
If s is empty, the function returns the empty list.
For example:

Test	Result
print(str_to_list('Programming'))
['P', 'R', 'G', 'M', 'N']
print(str_to_list('PROGRAMMING'))
['P', 'R', 'G', 'M', 'N']
print(str_to_list(''))
[]
print(str_to_list('1234543'))
['5', '3', '1']
print(str_to_list(1234543))
not a string
print(str_to_list('359'))
['9', '5', '3']
print(str_to_list('3a9'))
alphanumeric



```python
# Questions 01:
from typing import Any, Dict, List, Union


def add_n_print(L: list, obj: object, pos: int, replace: bool) -> Union[list, str]:

    if not isinstance(L, list):
        return "not a list"
    
    min_valid_position = 0
    max_valid_position = len(L)

    if not (min_valid_position <= pos <= max_valid_position):
        return "invalid pos"
    
    L_copy = L.copy()

    if not replace:
        L_copy.insert(pos, obj)
    else:
        L_copy.insert(pos, obj)
        if not (pos == len(L)):
            L_copy.pop(pos + 1)

    return L_copy


data = [
    {
        "value" : (99, 7 , 0, False),
        "answer": "not a list",
    },
    {
        "value" : ([99], 7, 0, False),
        "answer": [7, 99],
    },
    {
        "value" : ([99], 7, 1, False),
        "answer": [99, 7],
    },
    {
        "value" : ([99], 7, 2, False),
        "answer": "invalid pos",
    },
    {
        "value" : ([1, 2, 3], "a", 2, False),
        "answer": [1, 2, "a", 3],
    },
    {
        "value" : ([1, 2, 3], "a", 2, True),
        "answer": [1, 2, "a"],
    },
    {
        "value" : ([1, 2, 3], "a", 3, False),
        "answer": [1, 2, 3, "a"],
    },
    {
        "value" : ([1, 2, 3], "a", 3, True),
        "answer": [1, 2, 3, "a"],
    },
]

for d in data:
    response = add_n_print(*d["value"])
    answer = d["answer"]
    print(f"Response: {response}")
    print(f"Answer: {answer}")
    print(f"Passed Test: {response == answer}")
    print("-" * 20)
```

```python
# Question 02:
vowels = ["A", "E", "I", "O", "U"]
odd_numbers = ["1", "3", "5", "7", "9"]

def str_to_list(s: str) -> list:
    
    if not isinstance(s, str):
        return "not a string"

    s_processed = list(s.upper())

    if not s_processed:
        return s_processed
    
    if s.isalpha():
        letter_collected = []
        for letter in s_processed:
            if letter in vowels:
                continue
            if letter not in letter_collected:
                letter_collected.append(letter)
        return letter_collected

    elif s.isnumeric():
        letter_collected = []
        for letter in s_processed:
            if letter in odd_numbers:
                if letter not in letter_collected:
                    letter_collected.append(letter)
        return letter_collected[::-1]
    
    else:
        return "alphanumeric"


data = [
    {
        "value" : "Programming",
        "answer": ['P', 'R', 'G', 'M', 'N'],
    },
    {
        "value" : 'PROGRAMMING',
        "answer": ['P', 'R', 'G', 'M', 'N'],
    },
    {
        "value" : "",
        "answer": [],
    },
    {
        "value" : '1234543',
        "answer": ['5', '3', '1'],
    },
    {
        "value" : 1234543,
        "answer": "not a string",
    },
    {
        "value" : '359',
        "answer": ['9', '5', '3'],
    },
    {
        "value" : '3a9',
        "answer": "alphanumeric",
    },
]

for d in data:
    print(str_to_list(d["value"]) == d["answer"])
```

```python
def is_even(n: int):
    return n % 2 == 0

data = [-1, 0, 111, 100]

for d in data:
    print(is_even(d))
```

<p style="font-family: 'Courier New', Courier, monospace;">
Write a function rent(days) that takes an integer number of days of rent and computes the total cost.

<p style="font-family: 'Courier New', Courier, monospace;">
The function always returns a float (if days &lt; 1, return 0.0).

        </tr>
        <tr>
            <td>print(rent(0))</td>
            <td>0.0</td>
        </tr>
        <tr>
            <td>print(rent(1))</td>
            <td>39.99</td>
        </tr>
        <tr>
            <td>print(rent(2))</td>
            <td>79.99</td>
        </tr>
        <tr>
            <td>print(rent(3))</td>
            <td>120.0</td>
        </tr>
        <tr>
            <td>print(rent(4))</td>
            <td>160.0</td>
        </tr>
        <tr>
            <td>print(rent(5))</td>
            <td>200.0</td>
        </tr>
        <tr>
            <td>print(rent(10))</td>
            <td>400.0</td>
        </tr>
    </tbody>
</table>


```python
extra_day = 40
defined_days = {
    1: 39.99,
    2: 79.99,
    3: 120,
    4: 160,
}

def rent(days: int) -> float:
    if days in defined_days:
        return float(defined_days[days]) 
    else:
        return float(defined_days[3] + (days - 3) * extra_day) if days > 0 else float(0)
    
data = [-2, 0, 1, 2, 3, 4, 5, 10]

for d in data:
    print(rent(d))
```
